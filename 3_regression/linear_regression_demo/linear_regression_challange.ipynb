```python
# imports
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# get data
dataframe = pd.read_csv('challenge_dataset.txt', header=None)
x = dataframe[[0]]
y = dataframe[[1]]
```

```python
# create model
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
# visualize the model
plt.scatter(x,y)
plt.plot( x, model.predict(x) )
plt.show()
```

```python
error = mean_squared_error( y, model.predict(x) )
print("Mean squared error: %.2f" % error)
```
